```python
import pandas as pd

data = {
    "Order_ID": [1001,1002,1003,1004,1005,1006,1007,1008,1009,1010,1011,1012,1013,1014,1015],
    "Product": ["Notebook","Laptop","Pen Set","Mouse","Keyboard","Printer","Marker Set",
                "Headphones","USB Drive","Scanner","Calculator","Webcam","Tablet","File Folder","Monitor"],
    "Category": ["Stationery","Electronics","Stationery","Accessories","Accessories","Electronics",
                 "Stationery","Accessories","Accessories","Electronics","Stationery",
                 "Accessories","Electronics","Stationery","Electronics"],
    "Quantity": [50,2,100,15,12,3,80,20,35,5,40,8,4,120,6],
    "Price": [120,45000,250,350,800,8500,180,1200,400,6500,350,2500,25000,150,12000],
    "Month": ["January","January","January","February","February","February","February",
              "March","March","March","April","April","April","April","April"]
}

df = pd.DataFrame(data)
df
```

```python
df["Total_Revenue"] = df["Quantity"] * df["Price"]
df
```

```python
df.groupby("Product")["Total_Revenue"].sum()
```
