```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "Employee_ID": [501,502,503,504,505,506,507,508,509,510,511,512,513,514,515],
    "Name": ["John","Sarah","Mike","Lisa","Tom","Emma","David","Rachel","Chris",
             "Anna","Peter","Julia","Mark","Sophie","Kevin"],
    "Department": ["IT","HR","IT","Finance","IT","HR","Finance","IT","HR",
                   "Finance","IT","HR","Finance","IT","Finance"],
    "Experience": [5,3,8,6,2,7,4,10,5,9,6,4,8,3,5],
    "Salary": [65,45,85,72,42,68,58,95,55,88,75,50,82,52,64],
    "Age": [28,26,32,30,24,31,27,35,29,34,30,28,33,25,29]
}

df = pd.DataFrame(data)
df
```

```python
plt.figure()
plt.hist(df["Salary"], bins=6)
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.title("Salary Distribution")
plt.show()
```

```python
plt.figure()
sns.boxplot(y=df["Age"])
plt.title("Age Distribution (Box Plot)")
plt.show()
```

```python
plt.figure()
df["Department"].value_counts().plot(kind="bar")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.title("Department-wise Employee Count")
plt.show()
```

```python
plt.figure()
plt.scatter(df["Experience"], df["Salary"])
plt.xlabel("Experience (Years)")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.show()
```

```python
plt.figure()
sns.heatmap(df[["Experience", "Salary", "Age"]].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
